```python
## 导入库
import random
import time

import pymysql


# 导入规范  PEP8规范
# 导入的第一个部分，就是最上边的部分，是我们导入的系统自带的一些库、函数，首先导入。
# 加上一个空行
# 第二个部分导入的是我们引入的三方库的导入信息，三方库就是我们通过pip install 安装的叫三方库
# 加上一个空行
# 第三个部分导入的是我们自己写的函数
```

```python
##  连接数据库
# 声明一个变量
db_info = {
    "host":"192.168.1.80", # 我们要连接的数据库的IP地址或者是域名，如果是本地的话可以写localhost或者是127.0.0.1
    "user":"admin", # 我们连接数据库的用户名
    "password":"123", # 我们连接数据库的密码
    "database":"management_systems", # 我们要连接的数据库的名字
    "charset":"utf8" # 这是字符集
}

# 开始连接
conn = pymysql.connect(**db_info)

# cursor是执行sql语句的对象
cursor = conn.cursor()
```

```python
## 简单的演示
sql = "select * from school"
cursor.execute(sql)
```

```python
## 查询时，获取数据
cursor.fetchall()
```

```python
# sql语句格式化的演示
sql = "select %s from student where class_id=%s and sex=%s" % ("student_name","1","1")
cursor.execute(sql)
cursor.fetchall()
```

```python
## 第二种格式化
params = {
    "class_id":1,
    "student_name":"student_name",
    "student_birthday":"student_birthday",
    "sex":1
}

sql = "select {student_name},{student_birthday} from student where class_id={class_id} and sex={sex}".format(**params)
cursor.execute(sql)
cursor.fetchall()
```

```python
## 插入数据，学生表
# insert into student(class_id,student_name,student_birthday,sex) values(......)
# 我们先来生成是国内日字段
a1 = (1980,1,1,0,0,0,0,0,0) # 这个代表生日的开始时间
a2 = (2000,12,31,23,59,59,0,0,0) # 这个代表生日的结束时间

# 进行时间上的转换
start = time.mktime(a1)
end = time.mktime(a2)
```

```python
# # 我们有36个班级
for i in range(1,37):
    # 我们假设每个班有50名学生
    for j in range(1,51):
        # 拼接姓名
        random_xing = random.randint(0,len(xing)-1)
        random_ming = random.randint(0,len(ming)-1)
        
        t = random.randint(start,end) # 在开始和结束的时间戳当中随机出来一个
        date_t = time.localtime(t) # 把时间戳转换成时间
        birthday = time.strftime("%Y-%m-%d", date_t) # 把时间转换成一个可以插入数据库的时间格式
        
        # 拼接sql
        sql = "insert into student(class_id,student_name,student_birthday,sex) values( " + str(i) +",'" + xing[random_xing] + ming[random_ming] + "','" + birthday +"',"+str(random.randint(0,1))+")"            
        print(sql)
        cursor.execute(sql)
        
```

```python
## 插入数据的时候，必须要有一个提交的动作，否则，数据将会保存在内存中，不会被写入到硬盘中
conn.commit()
```

```python
## 要插入学生的分数就需要查询出学生的id和课程的id，并且最终插入的分数不能超过满分，所以我们还得查询出这个课程的满分是多少
get_student_info_sql = """
select s.id,cc.id,cc.full_mark
from grade g
	join class c
		on g.id = c.grade_id
	join student s
		on c.id = s.class_id
	join course cc
		on g.id = cc.grade_id

"""
cursor.execute(get_student_info_sql)
result = cursor.fetchall()
result
```

```python
## 插入分数
for i in range(len(result)):
    insert_values = result[i]
    score = random.randint(0, insert_values[2])
    sql = """
        insert into score(student_id,course_id,score) values (%s,%s,%s)
    """ % (insert_values[0], insert_values[1], score)
    cursor.execute(sql)

conn.commit()
```

```python
## 一个优秀的工程师，一定会在操作完数据库之后做以下的动作
# 关闭链接
cursor.close()
conn.close()
# 这是一个良好的操作习惯
# 这样做的原因是，数据库的连接是有限的，如果我们不主动断开连接，那么这个连接就会一直被占用
# 当连接满了，别人就连不上了，就会报错
```
