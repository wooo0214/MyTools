## 区域分布地图


```python
from pyecharts.charts import Map
# 为了保持下边的地理分布图使用的省份一致，我们需要处理一下数据，这里先处理一下自治区
def strip_region(iterable):
    result = []
    for i in iterable:
        if i.endswith('自治区'):
            if i == '内蒙古自治区':
                i = i[:3]
                result.append(i)
            else:
                result.append(i[:2])
        else:
            result.append(i)
    return result
```

```python
## 处理自治区
data_address.index = strip_region(data_address.index)

# 处理省，把末尾的’省‘字去掉
data_address.index = data_address.index.str.strip('省')
data_address
```

```python
# 把数据编程地图里要求的格式
data_map = [list(z) for z in data_address.items()]
```

```python
m = (
    Map()
    .add('订单分布地图', data_map,'china')
)
m.render_notebook()
```

##  订单各环节转化漏斗图


```python
## 总订单数
total_order_count = data['订单编号'].size
total_order_count
```

```python
## 付款订单数
pay_order_count = data['订单付款时间'].notnull().sum()
pay_order_count
```

```python
## 到款订单数
completed_deal_count = (data['买家实际支付金额']!=0.0).sum()
completed_deal_count
```

```python
data.head()
```

```python
##  全额到款订单数
full_pay_order_count = (data['退款金额']==0.0).sum()-data['订单付款时间'].isnull().sum()
full_pay_order_count
```

```python
order_count_data = pd.Series(
    {
        "总订单数":total_order_count,
        "付款订单数":pay_order_count,
        "到款订单数":completed_deal_count,
        "全额到款订单数":full_pay_order_count
    },name='订单数'
).to_frame()
order_count_data
```

```python
from pyecharts.charts import Funnel
name = '订单数'
f = Funnel().add(series_name=name,
                 data_pair=[list(z) for z in zip(order_count_data.index,order_count_data[name])]
                ).set_global_opts(title_opts=opts.TitleOpts(title='订单转化'))
f.render_notebook()
```

```python
import pandas as pd
data = pd.read_table('ad_performance.txt',delimiter='\t')
data.head()
```

```python
## 广告卖点数据绘制词云图
## 数据形式[("名字",数量),()]

word_data = data.groupby('广告卖点').count()['渠道代号']
word_data
```

```python
word_cloud_data = [(x,int(y)) for x,y in zip(word_data.index,word_data.values)]
word_cloud_data
```

```python
## 绘制词云图
from pyecharts import options as opts
from pyecharts.charts import WordCloud

w = WordCloud().add('',word_cloud_data).set_global_opts(title_opts=opts.TitleOpts(title='卖点词云图'))
w.render_notebook()
```

```python
from sklearn.cluster import KMeans
x1 = data[['日均UV','平均注册率','平均搜索量','访问深度','订单转化率','投放总时间']]
inertia = []


for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit(x1)
    inertia.append(km.inertia_)
l = (
    Line(init_opts=opts.InitOpts(width='1000px', height='500px'))
    .add_xaxis(range(1,11))
    .add_yaxis('', inertia)
)
l.render_notebook()
```

```python
km = KMeans(n_clusters=5)
y_means = km.fit_predict(x1)
y_means
```

```python
##  x1 = data[['日均UV','平均注册率','平均搜索量','访问深度','订单转化率','投放总时间']]

type1_count = x1[y_means==0]['日均UV'].count()
type2_count = x1[y_means==1]['日均UV'].count()
type3_count = x1[y_means==2]['日均UV'].count()
type4_count = x1[y_means==3]['日均UV'].count()
type5_count = x1[y_means==4]['日均UV'].count()


type1 = [[
    x1[y_means==0]['日均UV'].sum()/type1_count,
    x1[y_means==0]['平均注册率'].sum()/type1_count,
    x1[y_means==0]['平均搜索量'].sum()/type1_count,
    x1[y_means==0]['访问深度'].sum()/type1_count,
    x1[y_means==0]['订单转化率'].sum()/type1_count,
    x1[y_means==0]['投放总时间'].sum()/type1_count,  
]]


type2 = [[
    x1[y_means==1]['日均UV'].sum()/type2_count,
    x1[y_means==1]['平均注册率'].sum()/type2_count,
    x1[y_means==1]['平均搜索量'].sum()/type2_count,
    x1[y_means==1]['访问深度'].sum()/type2_count,
    x1[y_means==1]['订单转化率'].sum()/type2_count,
    x1[y_means==1]['投放总时间'].sum()/type2_count,  
]]
type3 = [[
    x1[y_means==2]['日均UV'].sum()/type3_count,
    x1[y_means==2]['平均注册率'].sum()/type3_count,
    x1[y_means==2]['平均搜索量'].sum()/type3_count,
    x1[y_means==2]['访问深度'].sum()/type3_count,
    x1[y_means==2]['订单转化率'].sum()/type3_count,
    x1[y_means==2]['投放总时间'].sum()/type3_count,  
]]
type4 = [[
    x1[y_means==3]['日均UV'].sum()/type4_count,
    x1[y_means==3]['平均注册率'].sum()/type4_count,
    x1[y_means==3]['平均搜索量'].sum()/type4_count,
    x1[y_means==3]['访问深度'].sum()/type4_count,
    x1[y_means==3]['订单转化率'].sum()/type4_count,
    x1[y_means==3]['投放总时间'].sum()/type4_count,  
]]
type5 = [[
    x1[y_means==4]['日均UV'].sum()/type5_count,
    x1[y_means==4]['平均注册率'].sum()/type5_count,
    x1[y_means==4]['平均搜索量'].sum()/type5_count,
    x1[y_means==4]['访问深度'].sum()/type5_count,
    x1[y_means==4]['订单转化率'].sum()/type5_count,
    x1[y_means==4]['投放总时间'].sum()/type5_count,  
]]
```

```python
from pyecharts.charts import Radar

c = (Radar()
    .add_schema(
        schema=[
            opts.RadarIndicatorItem(name='日均UV'),
            opts.RadarIndicatorItem(name='平均注册率'),
            opts.RadarIndicatorItem(name='平均搜索量'),
            opts.RadarIndicatorItem(name='访问深度'),
            opts.RadarIndicatorItem(name='订单转化率'),
            opts.RadarIndicatorItem(name='投放总时间'),
        ]
    )
    .add('1类用户',type1)
    .add('2类用户',type2)
     .add('3类用户',type3)
     .add('4类用户',type4)
     .add('5类用户',type5)
     .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
     .set_global_opts(legend_opts=opts.LegendOpts(selected_mode='single'),
                     title_opts=opts.TitleOpts(title='各类型客户'))
    )
c.render_notebook()
```

```python
## 柱状图与折线图叠加显示

x_axis = list(data['渠道代号'].head(10).values)
y1_axis = list((data['平均注册率']*100).head(10))
y2_axis = list((data['订单转化率']*100).head(10))
y3_axis = list((data['平均搜索量']*100).head(10))



```

```python
bar = (
    Bar()
    .add_xaxis(x_axis)
    .add_yaxis('平均注册率',y1_axis)
    .add_yaxis('订单转化率',y2_axis)
    .extend_axis(
        yaxis=opts.AxisOpts(
            axislabel_opts = opts.LabelOpts(formatter='{value}个')
        )
    )
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(
        title_opts=opts.TitleOpts('折线图与柱状图叠加'),
        yaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(formatter='{value}%')),
        datazoom_opts=[
            opts.DataZoomOpts(yaxis_index=0),
            opts.DataZoomOpts(type_='inside',yaxis_index=0)
            
        ]
    )
)

line=Line().add_xaxis(x_axis).add_yaxis('搜索量',y3_axis,yaxis_index=1)
bar.overlap(line)
bar.render_notebook()
```
