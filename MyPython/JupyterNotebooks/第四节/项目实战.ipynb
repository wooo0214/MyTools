## 旅游网站

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('旅游.csv',index_col=0)
data.head()
```

```python
## 提取酒店评分
data['酒店评分']=data['酒店'].str.extract('(\d\.\d)分/5分')
data.head()
```

```python
# 提取酒店等级
data['酒店等级']=data['酒店'].str.extract(' (.+) ')
data.head()
```

```python
data['天数']=data['路线名'].str.extract('(\d)天\d晚')
data.head()
```

## 租房数据提取

```python
zufang = pd.read_csv('租房.csv')
zufang.head()
```

```python
# 使用正则表达式提取网址
zufang['url'] = zufang['house_title'].str.extract('(https://.*)')
zufang.head()
```

```python
## 拼接几室几厅几卫
def str_concat(a,b,c):
    return str(a) + '室' + str(b) + '厅' + str(c) + '卫'
zufang['戶型'] = zufang.apply(lambda x : str_concat(x['bedroom_num'],x['hall_num'],x['bathroom_num']), axis=1)
zufang.head()
```

## 评论数据项目实战

```python
pinglun = pd.read_csv('评论.csv')
pinglun.head()
```

```python
## 用户活跃时间段分析
time = pinglun['creationTime'].str.extract(' (.*)')
time
```

```python
pinglun['hour'] = time[0].str.extract('^(\d{2}):')
pinglun.head()
```

```python
# 柱状图绘制，关注哪一个时间段更活跃
a = np.sort(pinglun['hour'].unique())
a
```

```python
height = pinglun.groupby('hour').count()['content']
height
```

```python
font = {'family':'SimHei','size':15}
plt.rc('font',**font)

plt.figure(figsize=(20,5))

plt.bar(a,height=height)
plt.title('用户评论活跃时间段')
gca = plt.gca()
gca.spines['right'].set_color('none')
gca.spines['top'].set_color('none')
```

```python
## 把句子变成词

import jieba # 如果要安装的话  pip install jieba
my = '我爱北京天安门，天安门上太阳升'

# 全模式
s = jieba.cut(my, cut_all=False)
r = '-'.join(s)
r
```

```python
# 精确模式
s = jieba.cut(my, cut_all=True)
r = '-'.join(s)
r
```

```python
s = jieba.lcut(my)
s
```

```python
recommend = pinglun['content']
recommend
```

```python
text = ''
for r in recommend:
    ## 去掉无用的词，停用词
    if r == " ":
        continue
    text += r

## 我们下边要绘制的词云图，需要用到一个字符串，里边包含很多词，词与词之间使用空格来进行分割的
data_cut = " ".join(jieba.lcut(text))

    
```

```python
## 绘制词云图
from wordcloud import WordCloud
# 解决文字显示问题
# 声明一个变量，用来存储本地字体的路径。 你可以到网上搜索，windows或者是mac字体存储路径
font = '/Library/Fonts/Microsoft/Kaiti.ttf'

# 生成词云图
w = WordCloud(font_path=font)
w.generate(data_cut)

# 把这个词云图写在一个图片里
image = w.to_file('评论.png')
# 显示这个图片
from PIL import Image
display(Image.open('评论.png'))
```
